# Dataset Prepration

```python
class BuildDataFrames: 
    
    def __init__(self, dataFrame_path, DataFrame_type): #, augmentation, oversample_method
        
        self.dataFrame_path = dataFrame_path
        self.DataFrame_type = DataFrame_type
        #self.augmentation = augmentation
        #self.ReadDataFrames()
  
    def ReadDataFrames(self):
        
        self.DataFrame = pd.read_csv(self.dataFrame_path)
        self.DataFrame.drop(['id', 'label'], axis=1, inplace=True)
        return self.DataFrame
    
    def Normalization(self, min_max_scaler_obj = None):
        
        if (self.DataFrame_type == 'train'):
            DataFrame = self.DataFrame
            numeric_col_list = DataFrame.select_dtypes(include='number').columns
            DataFrame_numeric_col = DataFrame[numeric_col_list].values
            min_max_scaler = preprocessing.MinMaxScaler()
            DataFrame_numeric_col_scaled = min_max_scaler.fit_transform(DataFrame_numeric_col)
            DataFrame_numeric_col_scaled = pd.DataFrame(DataFrame_numeric_col_scaled, columns= numeric_col_list)
            self.DataFrame[numeric_col_list] = DataFrame_numeric_col_scaled
            return self.DataFrame, min_max_scaler
        elif (self.DataFrame_type == 'test'):
            DataFrame = self.DataFrame
            numeric_col_list = DataFrame.select_dtypes(include='number').columns
            DataFrame_numeric_col = DataFrame[numeric_col_list].values
            DataFrame_numeric_col_scaled = min_max_scaler_obj.transform(DataFrame_numeric_col)
            DataFrame_numeric_col_scaled = pd.DataFrame(DataFrame_numeric_col_scaled, columns= numeric_col_list)
            self.DataFrame[numeric_col_list] = DataFrame_numeric_col_scaled
            return self.DataFrame
        
    def OneHotEncoding(self, label_feature_name): # label_feature_name for UNSW-NB15 is 'label' for binary classification and 'attack_cat' for multi classification
        
        categorical_column_list = list(self.DataFrame.select_dtypes(include='object').columns)
        categorical_column_list.remove(label_feature_name)
  
        one_hot_DataFrame = pd.get_dummies(self.DataFrame, columns=categorical_column_list)
        one_hot_DataFrame_cols = one_hot_DataFrame.columns
        
        train_label_col = one_hot_DataFrame.pop(label_feature_name)
        one_hot_DataFrame.insert(len(one_hot_DataFrame.columns), label_feature_name, train_label_col) #position parameter for insert function starts with 0 so we dont need to len(one_hot_train.columns)+1 (after pop function)

        self.DataFrame = one_hot_DataFrame

        return self.DataFrame, one_hot_DataFrame_cols
    
    def RescalingBetweenTrainTest(self, train_one_hot_cols, test_one_hot_cols):
        
        if self.DataFrame_type == 'train':
            diff_test_train = list(set(test_one_hot_cols) - set(train_one_hot_cols))
            if len(diff_test_train)>0 :
                zero_values = [0] * len(self.DataFrame)
                for col in diff_test_train:
                    self.DataFrame[col] = zero_values
                    
                #upcoming 2 lines of code, take attack_cat column and put it at the last column
                train_label_col = self.DataFrame.pop('attack_cat')
                self.DataFrame.insert(len(self.DataFrame.columns), 'attack_cat', train_label_col)
                return self.DataFrame
            else:
                return self.DataFrame
        
        elif self.DataFrame_type == 'test':
            diff_train_test = list(set(train_one_hot_cols) - set(test_one_hot_cols))
            if len(diff_train_test)>0 :
                zero_values = [0] * len(self.DataFrame)
                for col in diff_train_test:
                    self.DataFrame[col] = zero_values
                
                #upcoming 2 lines of code, take attack_cat column and put it at the last column
                test_label_col = self.DataFrame.pop('attack_cat')
                self.DataFrame.insert(len(self.DataFrame.columns), 'attack_cat', test_label_col)
                return self.DataFrame
            else:
                return self.DataFrame

    def LabelBinarize(self, LabelBinarizerObj = None):
        
        if self.DataFrame_type == 'train':     
            LabelBinarizerObject_fittedOnTrainLabel = LabelBinarizer().fit(self.DataFrame['attack_cat']) # create an object of label binarizer, then fit on train labels
            TrainBinarizedLabel = LabelBinarizerObject_fittedOnTrainLabel.transform(self.DataFrame['attack_cat']) # transform train labels with that object
            TrainBinarizedLabelDataFrame = pd.DataFrame(TrainBinarizedLabel, columns= LabelBinarizerObject_fittedOnTrainLabel.classes_) #convert transformed labels to dataframe
            self.DataFrame = pd.concat([self.DataFrame.drop(['attack_cat'], axis=1), TrainBinarizedLabelDataFrame],axis=1) # concatinate training set after drop 'label' with created dataframe of binarized labels
            return self.DataFrame, LabelBinarizerObject_fittedOnTrainLabel
        
        elif self.DataFrame_type == 'test':
            
            TestBinarizedLabel = LabelBinarizerObj.transform(self.DataFrame['attack_cat'])
            TestBinarizedLabelDataFrame = pd.DataFrame(TestBinarizedLabel, columns= LabelBinarizerObj.classes_)
            self.DataFrame = pd.concat([self.DataFrame.drop(['attack_cat'], axis=1), TestBinarizedLabelDataFrame],axis=1)
            return self.DataFrame
            
```

```python
def SortColumnsBetweenTrainTest(train, test):
    train_cols = train.columns
    test_sortedBasedOnTrain = pd.DataFrame(columns= train_cols)
    for col in test_sortedBasedOnTrain:
        test_sortedBasedOnTrain[col] = test[col]
        
    return train, test_sortedBasedOnTrain

train_sorted, test_sorted = SortColumnsBetweenTrainTest(train_LabelBinerized, test_LabelBinerized)
```

```python
def SaveDataFrames(DataFrame, DataFrameType, version):
    DataFrame.to_csv(str(version)+ '_'+ DataFrameType+ '.csv',index=False)
    print('saved')
    
def SaveArray(X, y, ArrayType):
    X_name = 'X_' + ArrayType + '.npy'
    np.save(X_name, X)
    y_name = 'y_' + ArrayType + '.npy'
    np.save(y_name, y)
```
